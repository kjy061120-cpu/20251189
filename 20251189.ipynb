```
import csv
from datetime import datetime
from heap import Heap  # 교재 파일

h = Heap()

# 1. CSV 읽기
with open('birthday.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)

    for row in reader:
        name = row['name']
        birthday = row['birthday']

        # 2. 날짜 변환
        date_obj = datetime.strptime(birthday, "%Y-%m-%d")

        # 3. heap에 넣기 (늦은 순 위해 음수)
        key = -date_obj.timestamp()
        h.insert((key, name, birthday))

# 4. 출력
print("생일이 늦은 순 10명:")

for i in range(10):
    if len(h.data) > 0:  # heap 내부 리스트 기준
        key, name, birthday = h.delete()
        print(f"{i + 1}. {name} - {birthday}")
```
