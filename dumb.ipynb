```
import csv
import heapq
from datetime import datetime


heap = []


with open('birthday.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    header = next(reader)

    for row in reader:
        name = row[0]
        birthday_str = row[1]

        birthday = datetime.strptime(birthday_str, "%Y-%m-%d")


        heapq.heappush(heap, (-birthday.timestamp(), name, birthday_str))

print("생일이 늦은 순서 상위 10명:")
for i in range(10):
    if heap:
        _, name, birthday_str = heapq.heappop(heap)
        print(f"{i+1}. {name} - {birthday_str}")
```
