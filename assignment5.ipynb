```
import csv
import heapq
from datetime import datetime


heap = []


with open('birthday.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    header = next(reader)

    for row in reader:
        name = row[0]
        birthday_str = row[1]

        birthday = datetime.strptime(birthday_str, "%Y-%m-%d")


        heapq.heappush(heap, (-birthday.timestamp(), name, birthday_str))

print("생일이 늦은 순서 상위 10명:")
for i in range(10):
    if heap:
        _, name, birthday_str = heapq.heappop(heap)
        print(f"{i+1}. {name} - {birthday_str}")
```

```
import csv

# 노드 클래스 (교재 구조 가정)
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

# 원형 이중 연결 리스트 클래스
class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            tail = self.head.prev

            tail.next = new_node
            new_node.prev = tail

            new_node.next = self.head
            self.head.prev = new_node

    def traverse(self):
        if self.head is None:
            return

        current = self.head
        while True:
            yield current.data
            current = current.next
            if current == self.head:
                break


# 리스트 생성
cdll = CircularDoublyLinkedList()

# CSV 읽기 및 저장
with open('birthday.csv', newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)

    for row in reader:
        try:
            name = row['name']
            birthday = row['birthday']
            group = row['group']

            # 생년월일이 없는 경우 예외 처리
            if birthday is None or birthday.strip() == "":
                raise ValueError("생년월일 없음")

            cdll.append({
                'name': name,
                'birthday': birthday,
                'group': group
            })

        except Exception as e:
            print(f"[경고] {row.get('name','이름없음')} 데이터 오류: {e}")


# 같은 조 선택 (예: A조)
target_group = "A"

print(f"\n[{target_group}조 친구들]")
for data in cdll.traverse():
    if data['group'] == target_group:
        print(f"{data['name']} - {data['birthday']}")
```

```
01. 최대 힙에서 가장 큰 원소가 항상 루트에 있다. 즉, 루트의 깊이가 가장 얕다. 임의의 최대 힙에서 더 얕은 곳에 있는 원소가 더 깊은 곳에 있는 원소보다 더 작은 값을 가질 수 있는가?
답 :  가능하다. 최대 힙의 조건은 부모 노드 >= 자식 노드 이므로 루트는 전체에서 가장 큰 값이다. 깊이와 값의 크기는 직접적인 관계가 없다.
```

```
02. 최대 힙 A[0…n-1]에서 A[0]는 항상 가장 큰 값을 갖고 있다. 반대로 마지막 원소인 A[n 1]은 항상 가장 작은 값을 갖는가?
답 : 아니다. 루트만 최대값을 보장하고 마지막 원소는 단지 리프 중 하나이기 때문이다.
```

```
03. 임의의 배열 A[0…n-1]을 힙으로 만드는 buildHeap() 알고리즘에서 총 n개의 원소 중 루트의 자격으로 스며내리기를 할지 알아보지 않고 그냥 넘어가는 원소(힙의 노드) 수는 정확하게 몇 개인가?
답 : [n/2]개 이다. 리프는 자식이 없으므로 스며내리기를 수행할 필요가 없다.
```

```
04. 배열 A[0…n-1]을 대상으로 하는 스며내리기 알고리즘에서 최악의 경우에 해당하는 시간과 최선의 경우에 해당하는 시간은 어떻게 되는가? Θ−표기로 나타내시오.
답 : 최악 : 루트에서 리프까지 내려가는 경우 Θ(log n), 최선 : 한 번 비교 후 종료 Θ(1)
```

```
06. 어떤 학생이 다음과 같은 질문을 했다. “buildHeap() 알고리즘에서는 아래쪽에서부터 시작해서 스며내리기를 반복하는데 만약 반대 방향으로 하면 어떤가요?(즉, 위쪽에서부터 시작해서 스며오르기를 반복)” 이렇게 해도 힙이 만들어진다. 이 방법은 본문에 제시한 방법에 비해 더 효율적인가? 점근적 시간으로 말하시오.
답 : 아니다. 더 비효율적이다. 기존 방식인 스며내리기는 Θ(n) 이고 제안 방식은  Θ(n log n)이므로 스며내리기 방식이 훨씬 효율적이다.
```
